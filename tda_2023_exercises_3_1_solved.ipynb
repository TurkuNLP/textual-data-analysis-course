## Load model


```python
MODEL = 'facebook/opt-1.3b'

model = transformers.AutoModelForCausalLM.from_pretrained(MODEL)
tokenizer = transformers.AutoTokenizer.from_pretrained(MODEL)
```

---

## Generation 

```python
def generate(prompt, max_new_tokens=10):
    inputs = tokenizer(prompt, return_tensors='pt')
    outputs = model.generate(inputs.input_ids, max_new_tokens=max_new_tokens)
    decoded = tokenizer.batch_decode(
        outputs,
        skip_special_tokens=True,
        clean_up_tokenization_spaces=False
    )
    output_text = decoded[0]
    generated = output_text[len(prompt):]
    return(f'**{prompt}**{generated}')
```

---

## Zero-shot experiments

(These are finicky, but can be made to work.)

```python
print(generate('''
The core emotions are Joy, Trust, Fear, Surprise, Sadness, Disgust, Anger, and Anticipation.
Which core emotions do the following sentences express?

Sentence: I'm not sure I have the courage to go out into the dark.

Answer:'''))
```
